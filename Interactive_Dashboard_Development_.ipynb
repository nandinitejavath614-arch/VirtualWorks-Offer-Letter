```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Sample Business Data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [15000, 18000, 22000, 25000, 27000, 30000],
    'Profit': [3000, 3500, 4200, 5000, 5500, 6200],
    'Orders': [120, 140, 180, 200, 230, 260]
}

df = pd.DataFrame(data)

# Dashboard Title
print("="*50)
print("INTERACTIVE BUSINESS DASHBOARD")
print("="*50)

# Sales Trend
fig1 = px.line(
    df,
    x='Month',
    y='Sales',
    title='Monthly Sales Trend',
    markers=True
)
fig1.show()

# Profit Trend
fig2 = px.bar(
    df,
    x='Month',
    y='Profit',
    title='Monthly Profit Analysis'
)
fig2.show()

# Orders Trend
fig3 = px.area(
    df,
    x='Month',
    y='Orders',
    title='Monthly Orders Growth'
)
fig3.show()

# KPI Metrics
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Orders'].sum()

print("\nKEY PERFORMANCE INDICATORS")
print(f"Total Sales : ₹{total_sales:,}")
print(f"Total Profit: ₹{total_profit:,}")
print(f"Total Orders: {total_orders}")

# Category Performance
category_data = pd.DataFrame({
    'Category':['Electronics','Clothing','Furniture','Books'],
    'Sales':[40000,25000,18000,12000]
})

fig4 = px.pie(
    category_data,
    values='Sales',
    names='Category',
    title='Sales Distribution by Category'
)
fig4.show()
```
