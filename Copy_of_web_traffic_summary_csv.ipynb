```python
import pandas as pd

data = {
    "User_ID": [101,102,103,104,105,106,107,108,109,110],
    "Page_Views": [5,8,2,10,6,3,12,4,7,1],
    "Session_Duration_Min": [12,18,4,25,15,6,30,8,20,2],
    "Bounce_Rate": [20,15,80,10,25,70,5,60,18,90],
    "Pages_Visited": [
        "Home>Products>Cart",
        "Home>Products>Checkout",
        "Home",
        "Home>Blog>Products>Checkout",
        "Home>Products",
        "Home",
        "Home>Blog>Products>Cart>Checkout",
        "Home>Blog",
        "Home>Products>Blog",
        "Home"
    ]
}

df = pd.DataFrame(data)

df.to_csv("web_traffic_data.csv", index=False)

print(df)
```

```python
import pandas as pd

df = pd.read_csv("web_traffic_data.csv")

print(df.head())
```

```python
print(df.info())
```

```python
print("Average Page Views:", df["Page_Views"].mean())
```

```python
print("Average Session Duration:",
      df["Session_Duration_Min"].mean())
```

```python
high_bounce = df[df["Bounce_Rate"] > 50]

print(high_bounce)
```

```python
for journey in df["Pages_Visited"]:
    print(journey.split(">")[-1])
```

```python
print(df["Pages_Visited"].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.bar(df["User_ID"], df["Page_Views"])
plt.title("Page Views by User")
plt.xlabel("User ID")
plt.ylabel("Page Views")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df["User_ID"],
         df["Session_Duration_Min"],
         marker="o")

plt.title("Session Duration")
plt.xlabel("User ID")
plt.ylabel("Minutes")
plt.show()
```

```python
summary = {
    "Average Page Views":
        df["Page_Views"].mean(),

    "Average Session Duration":
        df["Session_Duration_Min"].mean(),

    "Average Bounce Rate":
        df["Bounce_Rate"].mean()
}

summary_df = pd.DataFrame([summary])

summary_df.to_csv("web_traffic_summary.csv",
                  index=False)

print(summary_df)
```
